```python
from phidown import CopernicusDataSearcher

searcher = CopernicusDataSearcher()

# Search for SLSTR land surface temperature products
results = searcher.query_by_filter(
    collection_name='SENTINEL-3',
    # product_type='S3SLSTR',
    #product_type='SL_2_LST',
    #cloud_cover_threshold=0,
    aoi_wkt='POLYGON((10 50, 20 50, 20 55, 10 55, 10 50))',  # Central Europe
    start_date='2021-07-01',
    end_date='2021-07-31',
    attributes={
        'instrumentShortName':'SLSTR'
    }
)

df = searcher.execute_query()
print(f"Number of results: {len(df)}")
# Display the first few rows of the DataFrame
searcher.display_results(top_n=15)
```
