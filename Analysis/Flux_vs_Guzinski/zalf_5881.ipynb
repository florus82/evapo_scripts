```python
import sys
sys.path.append('/home/potzschf/repos/')
from helperToolz.helpsters import *
from helperToolz.dicts_and_lists import *
from helperToolz.guzinski import * 
from shapely.geometry import Point
import geopandas as gpd
```

```python
# station coordinates epsg:4326 (51.879095, 14.040279)

points = gpd.GeoDataFrame(
    geometry=[Point(14.040279, 51.879095)],
    crs="EPSG:4326"
)
```

```python
results = []


for comp, vrt_path in zip(['maxLST', 'minVZA'], ['/data/Aldhani/eoagritwin/et/Sentinel3/LST/SharpEvap/Brandenburg/FirstShot/evap/maxLST/vrt/',
                                            '/data/Aldhani/eoagritwin/et/Sentinel3/LST/SharpEvap/Brandenburg/FirstShot/evap/minVZA/vrt/']):
    for dir in dirfinder(vrt_path):
        if dir in ['Canopy_calc', 'Soil_calc']:
            continue
        else:
            for file in getFilelist(f'{vrt_path}{dir}', 'vrt'):
                with rasterio.open(file) as src:
                    point_reprj = points.to_crs(src.crs)
                    val = next(src.sample([(point_reprj.geometry.x, point_reprj.geometry.y)]))[0]

                    results.append({
                                'comp': comp,
                                'dir': dir,
                                'date': file.split(f'{dir}_')[-1].split('.')[0],
                                'val': val
                            })

# convert and export
df = pd.DataFrame(results)
df.to_csv(f'zalfID5881extract_FirstShot.csv', index=False)
```
